```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = """John went to the market. He bought apples.
Mary saw John there. She greeted him.
The dog was barking loudly. It was hungry."""
```

```python
sentences = sent_tokenize(text)

```

```python
male = ['john', 'ram', 'rohit']
female = ['mary', 'sita', 'priya']
```

```python
persons = []
objects = []

resolved_sentences = []

```

```python
male = ["john", "ram", "alex"]
female = ["mary", "sita", "anna"]

for s in sentences:
    words = word_tokenize(s)
    tagged = pos_tag(words)

    new_sentence = []

    for i, (w, t) in enumerate(tagged):
        lw = w.lower()

        if t in ["NNP", "NNPS"]:
            persons.append(lw)
            new_sentence.append(w)

        elif t == "NN":
            objects.append(lw)
            new_sentence.append(w)

        elif t == "PRP":
            if lw in ["he", "him"]:
                replacement = next((x for x in reversed(persons) if x in male), w)
                new_sentence.append(replacement)

            elif lw in ["she", "her"]:
                replacement = next((x for x in reversed(persons) if x in female), w)
                new_sentence.append(replacement)

            elif lw == "it":
                replacement = objects[-1] if objects else w
                new_sentence.append(replacement)

            else:
                new_sentence.append(w)

        else:
            new_sentence.append(w)

    resolved_sentences.append(" ".join(new_sentence))
```

```python
final_text = " ".join(resolved_sentences)
print(final_text)
```
